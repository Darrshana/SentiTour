### Module 2 - Vector Generation

### Set seed words

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)

positive_seed = ['good','amazing','beautiful','serene','love','enjoy','nice','peace','wonderful','stunning','splendid','delicious','yummy']
negative_seed = ['bad','worst','dirty','unsafe','horrible','hate','untidy','unclean','unworthy','ruined','theft','rob','headache','unhygenic']

def positive_tweets(tweet) :
    for i in range(0,len(positive_seed)):
        if positive_seed[i] in tweet.lower():
            tweet = re.sub(r'\[',"",tweet)
            tweet = re.sub(r'\]',"",tweet)
            tweet = re.sub(r'\"',"",tweet)
            tweet = re.sub(r'\'',"",tweet)
            return tweet
    return ""
def negative_tweets(tweet) : 
    for i in range(0,len(negative_seed)):
        if negative_seed[i] in tweet.lower():
            tweet = re.sub(r'\[',"",tweet)
            tweet = re.sub(r'\]',"",tweet)
            tweet = re.sub(r'\"',"",tweet)
            tweet = re.sub(r'\'',"",tweet)
            return tweet
    return ""
    
```

### Vector Space Generation

```python
def vector_space(wordlist):
    rows, cols = (2,len(wordlist)) 
    m = [[1 for i in range(cols)] for j in range(rows)]

    for i in range(0,int(len(wordlist))):
        if(wordlist[i].lower() in positive_seed) :
            m[0][i] = 1
            m[1][i] = 0
        elif (wordlist[i].lower() in negative_seed): 
            m[0][i] = 0
            m[1][i] = 1
        else:  
            #if not present in both consider it neutral
            m[0][i]=0.5
            m[1][1]=0.5
            
    pd.DataFrame(m, columns=wordlist)
    tf = TfidfTransformer()
    m = tf.fit_transform(m).todense()
    print(pd.DataFrame(m, columns=wordlist))
```
